### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

### 1.4 Удалите признаки date, zipcode, lat, long

```python
newHouseTable = pd.read_csv('kc-house-data.csv', encoding='1251')
```

```python
newHouseTable.head(5)
```

```python
price_per_sq_lot = []

for i, row in newHouseTable.iterrows():
    price_per_sq_lot.append(round(row['price']/ row['sqft_lot'],2))

price_per_sq_lot[:10]
```

```python
newHouseTable['price_per_sq_lot'] = price_per_sq_lot
newHouseTable.head(5)
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
delta_renovated = []

for i, row in newHouseTable.iterrows():
    if (row['yr_renovated']==0):
        delta_renovated.append(0)
    else:
        delta_renovated.append(row['yr_renovated'] - row['yr_built'])

delta_renovated[:10]
```

```python
newHouseTable['delta_renovated'] = delta_renovated
newHouseTable.head(5)
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
def yearSale(row):
    row = row.split()
    for el in row:
        year=int(el[0:4])
    return int(year)

newHouseTable['year_sale'] = newHouseTable['date'].apply(yearSale)
newHouseTable.head(5)
```

```python
def monthSale(row):
    row = row.split()
    for el in row:
        month=int(el[4:6])
    return int(month)

newHouseTable['month_sale'] = newHouseTable['date'].apply(monthSale)
newHouseTable.head(5)
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
newHouseTable.drop(columns=['date','lat','long'], inplace=True)
newHouseTable.head(5)
```
